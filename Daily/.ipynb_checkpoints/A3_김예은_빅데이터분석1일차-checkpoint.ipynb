```python
df = pd.DataFrame({'ABC' : ['A','B','C','A'],
                  'abc' : ['a','b','c','c'],
                  'group1' : ['G1','G2','G3','G4'],
                  'value1' : [11, 13, 12, 14],
                  'val1_mod' : [0.231 , 0.197, 0.213, 0.199]})
df
```

```python
df.filter(items = ['group1','value1'])
```

```python
#정규표현식 사용해서 열의 이름에 특정 문자 포함된 애들만 불러오기, '1'포함
df.filter(regex = '1', axis = 1)
```

```python
#1로 끝나는 columns 불러오기
df.filter(regex = '1$', axis = 1)
```

```python
#'a'가 들어가있는 열 불러오기
df.filter(like = 'a', axis = 1)
```

```python
#대문자 또는 소문자 a 가 들어 있는 열 (column)불러오기
df.filter(regex = '[Aa]', axis = 1)
```

```python
#12보다 큰 값만 찾아서 mask 하기
df.mask(df.value1 > 12)
```

```python
#a인 것만 mask
df.mask(df.abc == 'a')
```

```python
#다른 값으로 바꿀 때 사용
df.mask(cond = df.value1 >12, other = 999)
```

### 데이터 결합

```python
#concat. 조건없이 단순히 dataframe 합치기 axis =1 일 경우 옆으로 , 기본값은 아래로 붙이기
concat(A,B)
A.append(B)
#join. outer는 합집합 , inner는 교집합
#ignore_index 기존 index 유지 시 False, 무시 시 True
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B2','B3','B4'],
                   'C':['C0','C2','C3','C4'],
                   'D':['D0','D2','D3','D4']},
                  index = [0,1,2,3])
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#세로로 결합하기
result = pd.concat([df1,df2,df3])
result
```

```python
#가로로 결합하기
result = pd.concat([df1,df2,df3], axis = 1)
result
```

```python
#새로운 column명 집어넣고 싶을때
result = pd.concat([df1,df2,df3], axis = 1, ignore_index = True)
result
```

```python
#데이터 변수명 설정
result.columns = ['col_'+str(i) for i in range(12)]
```

```python
result
```

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']},
                  index = [0,1,2,3])
right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K0','K0','K0'],
                   'A':['C0','C1','C2','C3'],
                   'B':['D0','D1','D2','D3']},
                  index = [0,1,2,3])
```

```python
left
```

```python
right
```

```python
result = pd.merge(left,right, on =['key1','key2'])
result
```

```python
#왼쪽 df 기준으로 있는 것에 대입
result = pd.merge(left,right,on=['key1','key2'],how = 'left')
result
```

```python
#오른쪽 df 기준으로 ㅣㅇㅆ는 것에 대입
result = pd.merge(left,right,on=['key1','key2'],how = 'right')
result
```

```python
#전체 다 합치기
result = pd.merge(left,right,on=['key1','key2'],how = 'outer')
result
```

```python
result = pd.merge(left,right,on=['key1','key2'],how = 'outer', indicator = True)
result
```

```python
left = pd.DataFrame({'key1': [1,2,3,4],
                    'A': ['A1','A2','A3','A4'],
                    'B' : ['B0','B1','B2','B3']})
right = pd.DataFrame({'C': ['C1','C2','C5'],
                    'D' : ['D1','D2','D3']},
                    index = [1,2,5])
```

```python
left
```

```python
right
```

```python
result = left.join(right)
result
```

```python
#오른쪽 기준으로
result = left.join(right, how='right')
result
```

```python
#교집합
result = left.join(right, how='inner')
result
```

```python
#합집합
result = left.join(right, how='outer')
result
```

```python
#인덱스 기준이 아닌 특정 기준을 원할때
result = left.join(right, on ='key1')
result
```

```python
#최근 3년간 cctv 증가율 비교
#증가율 = 2016~2018 cctv 수/ 2016년 이전 cctv 수
cctv_seoul['최근증가율'] = (cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년'])/(cctv_seoul['cctv수']-(cctv_seoul['2018년']+cctv_seoul['2017년']+cctv_seoul['2016년']))*100

cctv_seoul.sort_values(by='최근증가율', ascending = False).head(5)
```
