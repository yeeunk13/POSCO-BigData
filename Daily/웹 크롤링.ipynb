```python
import requests
from bs4 import BeautifulSoup

def get_per_pbr(code) :
    url = f'https://finance.naver.com/item/main.nhn?code={code}'
    resp = requests.get(url)
    html = resp.text
    
    soup = BeautifulSoup(html, 'html5lib')
    per = soup.select('#_per')[0]
    pbr = soup.select('#_pbr')[0]
    
    per = float(per.text)
    pbr = float(pbr.text)
    return(per, pbr)
```

```python
get_per_pbr('005930')
```

```python
codes = ['000020','005930','035420','005490']
for code in codes:
    per, pbr = get_per_pbr(code)
    print(per, pbr)
```

```python
import requests
from bs4 import BeautifulSoup

def get_reviews(code) :
    url = f'https://movie.naver.com/movie/bi/mi/review.nhn?code={code}'
    resp = requests.get(url)
    html = resp.text
    
    soup = BeautifulSoup(html, 'html5lib')
    reviews =soup.select('#reviewTab > div > div > ul > li > a')
    for review in reviews[:4]:
        print(review.text.strip())
```

```python
from konlpy.tag import Hannanum
from collections import Counter
tagger = Hannanum()

text = '대한민국은 민주공화국이다. 대한민국의 주권은 국민에게 있고, 국민은 그 대표자나 국민투표에 의하여 주권을 행사한다.'
result = tagger.morphs(text)

stop_words = ['은','.','이','다','의','에게','있','고','나','그','에','어','을','하','ㄴ다',',']
result = [x for x in result if x not in stop_words]

c = Counter(result)
print(c.most_common(5))
```

### 텍스트 전처리

```python
from konlpy.tag import Okt
import re

#분석할 텍스트 읽어오기
with open('/home/piai/Downloads/news.txt', 'r', encoding = 'utf-8') as f:
    content = f.read()

#불필요한 심볼 없애기
p = re.compile('[\Wa-zA-Z0-9_]+')
content = re.sub(p, "", content)

#형태소 분석 및 단어 추출
tagger = Okt()
morphs = tagger.pos(content)
print(morphs)
```

```python
#명사만 추출하기
words = []
for word, pos in morphs:
    if pos == 'Noun':
        words.append(word)
print(words)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
data = pd.read_csv('/home/piai/Downloads/movies_metadata.csv', low_memory = False)

#데이터 읽은 후 20000개 샘플만 읽기
data = data.iloc[:20000]
data.head()
```

```python
#overview가 없는 데이터 처리
data['overview'] = data['overview'].fillna('')
print(data.shape)
```

```python
#data['title'].drop_duplicates()
```

```python
#TF-IDF
tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = tfidf.fit_transform(data['overview'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
#cosine similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim[0][:10])

title = 'The Dark Knight Rises'
title2index = pd.Series(data = data.index, index = data['title']).drop_duplicates()
index = title2index[title]
```

```python
title2index
```

```python
#index 번째 영화에 대한 유사도 값에 대해 [(0,유사도),(1,유사도) 형태로 저장
sim_scores = list(enumerate(cosine_sim[index]))
```

```python
#정렬하기
sim_scores.sort(key = lambda x:x[1], reverse = True) #내림차순 정렬
top10 = sim_scores[1:11]
top10_index = [x[0] for x in top10]
print(top10_index)
print(data['title'].iloc[top10_index])
```
