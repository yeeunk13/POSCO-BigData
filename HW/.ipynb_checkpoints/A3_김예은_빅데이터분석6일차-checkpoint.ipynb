### step0. 패키지 불러오기

```python
#패키지 불러오기 
import pandas as pd 
#데이터 시각화
import matplotlib.pyplot as plt
import matplotlib
#데이터 scaling
from sklearn.preprocessing import StandardScaler
#데이터 분할;train,test
from sklearn.model_selection import train_test_split
#분류SVM
from sklearn.svm import SVC
#최적 모델, 파라미터 검색
from sklearn.model_selection import GridSearchCV
#모델 성능 평가
from sklearn.metrics import confusion_matrix
```

### step1. 데이터 구성하기

```python
#데이터 구성하기
df_raw=pd.read_csv("/home/piai/Downloads/유방암.csv")
df_raw.head()
```

```python
#결측치 확인하기
df_raw.isnull().sum(axis = 0)
```

```python
#목표변수, 설명변수 데이터 지정
df_raw_y = df_raw['diagnosis']
df_raw_x = df_raw.drop('diagnosis', axis = 1 , inplace = False)
```

```python
#get_dummies: 데이터의 문자형 변수에 대한 더미 변수 생성
df_raw_dummy=pd.get_dummies(df_raw)
#더미변수 생성된 데이터의 상위 5개 row확인
df_raw_dummy.head()
```

```python
#데이터 구성하기 -데이터 분할
df_raw_y = df_raw_dummy["diagnosis"]
df_raw_x = df_raw_dummy.drop("diagnosis", axis=1)


df_train_x, df_test_x, df_train_y, df_test_y = train_test_split(df_raw_x, df_raw_y,test_size = 0.3, random_state = 1234)
df_train_x, df_val_x, df_train_y , df_val_y = train_test_split(df_train_x, df_train_y, test_size = 3/7, random_state = 1234)
print(f'train data X size :{df_train_x.shape}')
print(f'train data Y size :{df_train_y.shape}')
print(f'test data X size :{df_test_x.shape}')
print(f'test data Y size :{df_test_y.shape}')
print(f'val data X size :{df_val_x.shape}')
print(f'val data Y size :{df_val_y.shape}')
```

### step2: SVC 모델 생성

```python
#SCV모델 생성
svm_uncustomized=SVC(random_state=1234)
svm_uncustomized.fit(df_train_x,df_train_y)
#train 데이터 셋 정확도
print("Accuracy on training set:{:.3f}".format(svm_uncustomized.score(df_train_x, df_train_y)))
#val 데이터 셋 정확도
print("Accuracy on val set:{:.3f}".format(svm_uncustomized.score(df_val_x,df_val_y)))
```

### step2: SVC 모델 Parameter 변경

```python
#모델 parameter 조정
#C; 규제에 대한 강도 변경
train_accuracy =[]; 
val_accuracy=[]

#lc; 규제 (0.01,0.1,1,10)
para_c=[5**c for c in range(-2,4)]
for v_C in para_c:
    svm=SVC(C=v_C,random_state=1234)
    svm.fit(df_train_x,df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    val_accuracy.append(svm.score(df_val_x,df_val_y))

#데이터 테이블로 저장 
df_accuracy_c=pd.DataFrame()
df_accuracy_c["C"]=para_c
df_accuracy_c["TrainAccuracy"]=train_accuracy
df_accuracy_c["valAccuracy"]=val_accuracy

#C별 정확도 테이블
df_accuracy_c.round(3)
```

```python
#C별 정확도 그래프로 확인
plt.plot(para_c,train_accuracy,linestyle="-",label="Train Accuracy")
plt.plot(para_c,val_accuracy,linestyle="--",label="val Accuracy")
plt.xlabel("C"); plt.ylabel("Accuracy")
plt.legend()
```

* train accuracy와 val accuracy가 동시에 높아지는 20을 C로 지정

- Gamma: 훈련 샘플이 미치는 영향의 범위를 변경

```python
#Gamma: 훈련 샘플이 미치는 영향의 범위를 변경
#train 및 test 정확도 결과 저장용
train_accuracy =[]; 
val_accuracy=[]

#lc; 규제 (0.01,0.1,1,10)
para_gamma=[5**gamma for gamma in range(-2,4)]
for v_gamma in para_gamma:
    svm=SVC(gamma=v_gamma,C=20 ,random_state=1234)
    svm.fit(df_train_x,df_train_y)
    train_accuracy.append(svm.score(df_train_x,df_train_y))
    val_accuracy.append(svm.score(df_val_x,df_val_y))
    
#데이터 테이블로 저장 
df_accuracy_gamma=pd.DataFrame()
df_accuracy_gamma["gamma"]=para_gamma
df_accuracy_gamma["TrainAccuracy"]=train_accuracy
df_accuracy_gamma["valAccuracy"]=val_accuracy

#C별 정확도 테이블
df_accuracy_c.round(3)
```

```python
#Gamma별 정확도 그래프로 확인
plt.plot(para_gamma,train_accuracy,linestyle="-",label="Train Accuracy")
plt.plot(para_gamma,val_accuracy,linestyle="--",label="val Accuracy")
plt.xlabel("gamma"); plt.ylabel("Accuracy")
plt.legend()
```

Gamma 변화는 SVM 모델의 정확도에 영향을 미치지 않는 것으로 보임

### step2: scale 변환

```python
#설명변수의 값의 구간을 동일하게 하기 위해 scale 진행

v_feature_name = df_train_x.columns
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_raw_x)
df_scaled = pd.DataFrame(df_scaled, columns = v_feature_name)
df_scaled.head()
```

```python
#scaled 된 데이터 train, test, val로 분할

df_scaled_train_x, df_scaled_test_x = train_test_split(df_scaled, test_size = 0.3, random_state = 1234)
df_scaled_train_x, df_scaled_val_x = train_test_split(df_scaled_train_x, test_size = 3/7, random_state = 1234)

print(f'train data X size :{df_train_x.shape}')
print(f'test data X size :{df_test_x.shape}')
print(f'val data X size :{df_val_x.shape}')
```

### step2: scale 조절 후 SVC 모델 재실행

```python
#scaled 후 분할된 데이터 accuracy 측정

svm_scaled = SVC(random_state = 1234)
svm_scaled.fit(df_scaled_train_x, df_train_y)

print("Accuracy on training set:{:.3f}".format(svm_scaled.score(df_scaled_train_x, df_train_y)))
print("Accuracy on val set:{:.3f}".format(svm_scaled.score(df_scaled_val_x, df_val_y)))
```

### step2: 새로운 SVC 모델 Parameter 변경

```python
train_accuracy = [] 
val_accuracy = []

# c: 규제(0.01, 0..1, 1, 10)
para_c = [5 ** c for c in range(-2, 4)]
for v_C in para_c:
    svm = SVC(C=v_C, random_state=1234)
    svm.fit(df_scaled_train_x, df_train_y)
    train_accuracy.append(svm.score(df_scaled_train_x, df_train_y))
    val_accuracy.append(svm.score(df_scaled_val_x, df_val_y))

# 데이터 테이블로 저장
df_accuracy_c = pd.DataFrame()
df_accuracy_c["C"] = para_c
df_accuracy_c["TrainAccuracy"] = train_accuracy
df_accuracy_c["valAccuracy"] = val_accuracy
df_accuracy_c.round(3)
```

```python
# 모델 정확도 그래프 확인
plt.plot(para_c, train_accuracy, linestyle="-", label="Train Accuracy")
plt.plot(para_c, val_accuracy, linestyle="--", label="val Accuracy")
plt.ylabel("accuracy"); plt.xlabel("C")
plt.legend()
```

* C가 1일 때 모델의 정확도가 많이 증가하므로 C를 1로 임의설정

```python
#새로운 SVC 모델 파라미터 변경
#Gamma 훈련 샘플이 미치는 영향의 범위를 변경 결과
train_accuracy=[];
val_accuracy=[]

#gamma: 규제(0.01 , 0.1 , 1, 10)
para_gamma=[10 **gamma for gamma in range(-2,4)]
for v_gamma in para_gamma:
    svm=SVC(gamma=v_gamma,C=1,random_state=1234)
    svm.fit(df_scaled_train_x,df_train_y)
    train_accuracy.append(svm.score(df_scaled_train_x,df_train_y))
    val_accuracy.append(svm.score(df_scaled_val_x,df_val_y))

#데이터 테이블로 저장
df_accuracy_gamma=pd.DataFrame()
df_accuracy_gamma['gamma']=para_gamma
df_accuracy_gamma['TrainAccuracy']=train_accuracy
df_accuracy_gamma['valAccuracy']=val_accuracy
# Gamma 별 정확도 테이블
df_accuracy_gamma.round(3)
```

### step2: 새로운 SVC 모델 평가

```python
# Gamma 별 정확도 그래프로 확인
plt.plot(para_gamma, train_accuracy, linestyle = "-", label = 'Train Accuracy')
plt.plot(para_gamma, val_accuracy, linestyle = '--', label = 'val Accuracy')
plt.ylabel('accuracy'); plt.xlabel('gamma')
plt.legend()
```

* gamma가 1이상이면 train 데이터의 모델은 과대적합이 된다. 따라서 gamma를 0.01로 설정

### step3: 결론 도출

```python
# 최종 모델
svc_final = SVC(gamma = 0.01, C = 1, random_state = 1234)
svc_final.fit(df_scaled_train_x, df_train_y)

# 평가
y_pred = svc_final.predict(df_scaled_test_x)

# train 데이터 셋 정확도
print("Accuracy on training set: {:.3f}" .format(svc_final.score(df_scaled_train_x, df_train_y)))
# test 데이터 셋 정확도
print("Accuracy on test set: {:.3f} \n" .format(svc_final.score(df_scaled_test_x, df_test_y)))
# confusion matrix
print("Confusion matrix: \n {}" .format(confusion_matrix(df_test_y, y_pred)))
```

* confusion matrix 를 통해 확인해본 결과 실제 유방암 양성 환자를 음성이라 판단하는 경우가 11,  
실제 유방암 악성인 사람을 양성이라 판단하는 경우가 1개이다.  
* 2종 오류를 범할 확률이 낮다는 점에서 안정적이라 할 수 있지만 1종 오류가 11건 존재하기에 완벽한 모델이라 할 수 없다.
